### Read data from lil homie star

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import math 
```

#### Some basic data about the agent

```python
wheel_circ = 21.2 # cm circumference of each wheel 
clics_per_rev = 20 # number of "clicks" per revolution from rotary encode 
dist_bet_wheels = 16 # cm distance between the right and left wheels
agent_size = 19 # cm dimension of agent (assume square, good approximation)
start_pos = np.zeros(2) # (x,y) starting position of the robot
start_orient = 0 # radians starting heading orientation of the robot wrt the x-axis
```

#### Opening the file with data collected during the agent "exploration"

```python
data_file_name = '../processing_read_control/comm_trial_19_w_vis.txt' # data file with sensing and movement data
data_file = open(data_file_name,"r+")
data = data_file.readlines()
data_file.close()
print('Total number of data lines read: ' + str(len(data)))
print('First line: ')
data[0:10]
```

#### Clean the random characters and empty lines from the data

```python
[line.split('\n')[0] for line in data[0:20] if len(line.split('\n')[0]) > 0]
```

```python
data_clean = [line.split('\n')[0] for line in data if len(line.split('\n')[0]) > 0]
print(len(data_clean))
```

#### Splitting up the controls, the wheel data, and the distance data

```python
data_control = [line for line in data_clean if 'Control' in line]
data_wheel = [line for line in data_clean if 'LDir' in line]
data_sense = [line for line in data_clean if 'S1' in line]
print(len(data_control))
print(len(data_wheel))
print(len(data_sense))
```

#### Cleaning up the control data

```python
control_time = [int(line.split(' : ')[0]) for line in data_control]
control_value = [line.split(' : Control ')[-1] for line in data_control]
print(control_time[0:10])
print(control_value[0:10])
```

```python
# controls come in way more often than movement or sensing data 
# take away duplicates 
index = 1
while index < len(control_time):
    if control_time[index] == control_time[index-1]:
        control_time.pop(index)
        control_value.pop(index)
    else:
        index += 1
print(control_time[0:10])
print(control_value[0:10])
print(len(control_time))
print(len(control_value))
```

```python
controls = {
    'time': np.array(control_time),
    'values': np.array(control_value)
}
print(controls.keys())
```

#### Cleaning up the wheel sensing data
(Lcount, Rcount)  ----> Movement

(+,-)   ---->  Forward movement

(-,-)   ---->  Leftward movement

(+,+)   ---->  Rightward movement

(-,+)   ---->  Backward movement

```python
data_wheel[0:10]
```

```python
wheel_time = [int(line.split(' : ')[0]) for line in data_wheel]
wheel_lcnt = [int(line.split(' | ')[1].split(': ')[-1]) for line in data_wheel]
wheel_rcnt = [int(line.split(' | ')[-1].split(': ')[-1]) for line in data_wheel]
print(wheel_time[0:10])
print(wheel_lcnt[0:10])
print(wheel_rcnt[0:10])
```

```python
# remove sensings where neither wheel turned
index = 0
while index < len(wheel_time):
    if wheel_lcnt[index] == 0 and wheel_rcnt[index] == 0:
        wheel_lcnt.pop(index)
        wheel_rcnt.pop(index)
        wheel_time.pop(index)
    else:
        index += 1
print(wheel_time[0:10])
print(wheel_lcnt[0:10])
print(wheel_rcnt[0:10])
```

```python
wheels = {
    'time': np.array(wheel_time),
    'l_cnt': np.array(wheel_lcnt),
    'r_cnt': np.array(wheel_rcnt)*-1 # flip the sign of the right wheel because it makes sense
}
print(wheels.keys())
```

```python
time_diff = wheels['time'][1:] - wheels['time'][0:-1]
plt.hist(time_diff, bins=100)
plt.title('Time between wheel counts')
plt.xlabel('Time (ms)')
plt.ylabel('Counts')
plt.xlim(0,500)
```

```python
plt.hist(wheel_circ*wheels['l_cnt']/clics_per_rev)
plt.title('Left wheel distance traveled between sensings')
plt.xlabel('Distance (cm)')
```

```python
plt.hist(wheel_circ*wheels['r_cnt']/clics_per_rev)
plt.title('Right wheel distance traveled between sensings')
plt.xlabel('Distance (cm)')
```

#### Organizing ultrasonic range sensing

```python
data_sense[0:10]
```

```python
range_time = [int(line.split(' : ')[0]) for line in data_sense]
range_front = [float(line.split(' | ')[0].split(': ')[-1].split(' cm')[0]) for line in data_sense]
range_left = [float(line.split(' | ')[1].split(': ')[-1].split(' cm')[0]) for line in data_sense]
range_back = [float(line.split(' | ')[2].split(': ')[-1].split(' cm')[0]) for line in data_sense]
range_right = [float(line.split(' | ')[3].split(': ')[-1].split(' cm')[0]) for line in data_sense]
print(range_time[0:10])
print(range_front[0:10])
print(range_left[0:10])
print(range_back[0:10])
print(range_right[0:10])
```

```python
# turn 0 valued ranges into np.nan values
for i in range(len(range_front)):
    if range_front[i] == 0.0:
        range_front[i] = np.nan
    if range_left[i] == 0.0:
        range_left[i] = np.nan
    if range_back[i] == 0.0:
        range_back[i] = np.nan
    if range_right[i] == 0.0:
        range_right[i] = np.nan
```

```python
ranges = {
    'time': np.array(range_time),
    'front': np.array(range_front),
    'left': np.array(range_left),
    'back': np.array(range_back),
    'right': np.array(range_right)
}
print(ranges.keys())
```

```python
time_diff = ranges['time'][1:] - ranges['time'][0:-1]
plt.hist(time_diff, bins=100)
plt.title('Time between range sensings')
plt.xlabel('Time (ms)')
plt.ylabel('Counts')
plt.xlim(0,500)
```

```python
plt.hist(ranges['front'])
plt.title('Front range sensings')
plt.xlabel('Distance (cm)')
```

```python
plt.hist(ranges['back'])
plt.title('Back range sensings')
plt.xlabel('Distance (cm)')
```

```python
plt.hist(ranges['right'])
plt.title('Right range sensings')
plt.xlabel('Distance (cm)')
```

```python
plt.hist(ranges['left'])
plt.title('Left range sensings')
plt.xlabel('Distance (cm)')
```

#### Find sensings that happened right after movements

```python
sense_move_vals = {}
for key in ranges.keys():
    sense_move_vals[key] = []
for i,time_wheel in enumerate(wheels['time']):
#     sort_sense_times = np.argsort(np.abs(ranges['time'] - time_wheel))
#     for sense_time_index in sort_sense_times:
#         if (ranges['time'][sense_time_index] - time_wheel) > 0:
#             break
    sense_time_index = np.argmin(np.abs(ranges['time'] - time_wheel))
    for key in ranges.keys():
        sense_move_vals[key].append(ranges[key][sense_time_index])
for key in sense_move_vals.keys():
    sense_move_vals[key] = np.array(sense_move_vals[key])
print(sense_move_vals.keys())
print(len(sense_move_vals['time']), len(wheels['time']))
```

```python
plt.hist(sense_move_vals['time'] - wheels['time'], bins=100)
plt.title('Time to sense after moving')
plt.xlabel('Time (ms)')
```

#### Visualize sensings

```python
for key in sense_move_vals.keys():
    if key != 'time':
        plt.plot(sense_move_vals[key])
```
